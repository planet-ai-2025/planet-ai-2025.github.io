```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return (x-3)**2

def grad_f(x):
  return 2*(x-3)

def gradient_descent(x0, lr, epochs):
  x = x0
  history =[x]

  for i in range(epochs):
    pred = x
    error = f(x)
    print(f"Step {i:2d} | Error: {error:.6f} | Prediction: {pred:.6f}")
    grad = grad_f(x)
    x = x - lr * grad
    history.append(x)
  return np.array(history)
```

```python
def plot_gradient_descent_path(f, trajectory, x_min=-1.0, x_max=6.0, n_points=200,
                               label="f(x) = (x-3)^2",
                               title="Gradient Descent with Error Tracking",
                               show=True):
    """
    Plot f(x) over [x_min, x_max] and overlay the gradient descent path given by `trajectory`.

    Parameters
    ----------
    f : callable
        Function mapping array-like -> array-like.
    trajectory : array-like
        Sequence of x-values visited by gradient descent.
    x_min, x_max : float
        Range for plotting f(x).
    n_points : int
        Number of points to sample for plotting f(x).
    label : str
        Legend label for the function curve.
    title : str
        Title for the plot.
    show : bool
        Whether to call plt.show().

    Returns
    -------
    float
        The final value of x in the trajectory.
    """


    trajectory = np.asarray(trajectory)
    xs = np.linspace(x_min, x_max, n_points)
    ys = f(xs)

    plt.plot(xs, ys, label=label)
    plt.scatter(trajectory, f(trajectory), color="red")
    plt.plot(trajectory, f(trajectory), "r--", label="Gradient Descent Path")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.title(title)
    if show:
        plt.show()

    print("Final value of x:", trajectory[-1])
    return float(trajectory[-1])
```

```python
x0 = 0.0
lr = 0.1
epochs = 15

trajectory = gradient_descent(x0, lr, epochs)
```

```python
plot_gradient_descent_path(f, trajectory)
```

```python
x0 = 0.0
lr = 0.01
epochs = 100

trajectory = gradient_descent(x0, lr, epochs)
plot_gradient_descent_path(f, trajectory)
```
