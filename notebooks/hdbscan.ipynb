```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import hdbscan

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
iris = load_iris()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(iris.data)
```

```python
dbscan = DBSCAN(eps=0.7, min_samples=5)
dbscan.fit(X_scaled)
labels = dbscan.labels_

plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('DBSCAN clusters')
plt.show()
```

```python
hdbscan_clusterer = hdbscan.HDBSCAN(min_cluster_size=5)
hdbscan_labels = hdbscan_clusterer.fit_predict(X_scaled)
labels = hdbscan_labels

plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('HDBSCAN clusters')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot DBSCAN clusters
axs[0].scatter(X_scaled[:,0], X_scaled[:,1], c=labels)
axs[0].set_title('DBSCAN')

# Plot HDBSCAN clusters
axs[1].scatter(X_scaled[:,0], X_scaled[:,1], c=hdbscan_labels)
axs[1].set_title('HDBSCAN')

plt.show()
```
